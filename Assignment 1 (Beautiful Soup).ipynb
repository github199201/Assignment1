## Ques:1 -  Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html, "html.parser")
titles = bs.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

## Ques:2-Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
import requests
from bs4 import BeautifulSoup
page=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
page
page.content
soup=BeautifulSoup(page.content)
Movie_Name=soup.find_all('td',class_="titleColumn")
Movie_Titles=[]
for i in Movie_Name:
    Movie_Titles.append(i.text.replace('\n','').strip())
Movie_Titles[:100]
```

```python
Rate=soup.find_all('td',class_="ratingColumn imdbRating")
Rating=[]
for i in Rate:
    Rating.append(i.text.replace('\n','').strip())
Rating[:100]
```

```python
import pandas as pd

Internship=pd.DataFrame({})
Internship["Movie Name& Year"]=Movie_Titles[:100]
Internship["Rating"]=Rating[:100]
Internship
```

## Ques:4-Write a python program to scrap book name, author name, genre and book review of any 5 books from‘www.bookpage.com’

```python
import requests
from bs4 import BeautifulSoup
page=requests.get('https://bookpage.com/reviews')
page
page.content
soup=BeautifulSoup(page.content)
Name=soup.find_all('h4',class_="italic")
Book_Titles=[]
for i in Name:
    Book_Titles.append(i.text.replace('\n','').strip())
Book_Titles
A_Name=soup.find_all('p',class_="sans bold")
Author_Name=[]
for j in A_Name:
    Author_Name.append(j.text.replace('\n','').strip())
Author_Name
Gen=soup.find_all('p',class_="genre-links hidden-phone")
Genre=[]
for k in Gen:
    Genre.append(k.text.replace('\n','').strip())
Genre    
Review=soup.find_all('p',class_="excerpt")
Book_Review=[]
for l in Review:
    Book_Review.append(l.text.replace('\n','').strip())
Book_Review   
```

```python
import pandas as pd
books=pd.DataFrame({})
books["Name"]=Book_Titles[:5]
books["Author Name"]=Author_Name[:5]
books["Genre"]=Genre[:5]
books["Review"]=Book_Review[:5]
books
```

## Ques:8-Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
import requests
from bs4 import BeautifulSoup
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YP0EQ44zY2w')
page
page.content
soup=BeautifulSoup(page.content)
Name=soup.find_all('p',class_="period-name")
Period=[]
for i in Name:
    Period.append(i.text.replace('\n','').strip())
Desc=soup.find_all('p',class_="short-desc")
Short_Desc=[]
for i in Desc:
    Short_Desc.append(i.text.replace('\n','').strip())
Temp=soup.find_all('p',class_="temp temp-low")
Temprature2=[]
for i in Temp:
    Temprature2.append(i.get_text().strip())
Temp=soup.find_all('p',class_="temp temp-high")
Temprature1=[]
for i in Temp:
    Temprature1.append(i.get_text().strip())
Tempratures=[]
Tempratures=Temprature2+Temprature1
Descp=soup.find_all('div',class_="col-sm-10 forecast-text")
Description=[]
for i in Descp:
    Description.append(i.text.replace('\n','').strip())
import pandas as pd

Weather=pd.DataFrame({})
Weather["Period"]=Period
Weather["Short Description"]=Short_Desc
Weather["Temprature"]=Tempratures
Weather["Description"]=Description[:9]
Weather    
```

## Ques:9-Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.


```python
import requests
from bs4 import BeautifulSoup
page=requests.get('https://internshala.com/internships')
page
page.content
soup=BeautifulSoup(page.content)
Name=soup.find_all('div',class_="heading_4_5 profile")
Title=[]
for i in Name:
    Title.append(i.text.replace('\n','').strip())
C_Name=soup.find_all('div',class_="heading_6 company_name")
Company=[]
for i in C_Name:
    Company.append(i.text.replace('\n','').strip())
Sal=soup.find_all('div',class_="item_body")
Salry=[]
for i in Sal:
    Salry.append(i.text.replace('\n','').strip())
Salry    
CTC=[]

for i in range(2,len(Salry),4):
    CTC.append(Salry[i])
    

Date=soup.find_all('div',class_="item_body")
Apply_Date=[]
for i in Date:
    Apply_Date.append(i.text.replace('\n','').strip())
Apply=[]

for i in range(3,len(Apply_Date),4):
    Apply.append(Apply_Date[i])
import pandas as pd

Internship=pd.DataFrame({})
Internship["Job Title"]=Title[:10]
Internship["Company Name"]=Company[:10]
Internship["CTC"]=CTC[:10] 
Internship["Apply date"]=Apply[:10]
Internship
```
